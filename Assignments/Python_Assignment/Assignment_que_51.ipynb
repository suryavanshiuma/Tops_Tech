7) How memory is managed in Python? \
Python manages memory automatically primarily through a private heap space, reference counting, and a generational garbage collector. This means developers do not need to manually allocate or deallocate memory.

Python has a built-in Memory Manager that manages memory allocation inside a private heap. All Python objects and data structures are stored in this private heap, which is not directly accessible to the programmer. Python automatically manages memory using reference counting and a garbage collector that removes unused or unreferenced objects.

In simple language: \
✔ Python has a Memory Manager  \
✔ Memory is allocated in a private heap    \
✔ Objects are stored there \
✔ Garbage Collector frees unused memory

29) Write a Python function to get the largest number, smallest num and sum of all from a list.

```python
num=[4,6,8,3,5,9,1]

def operations(my_list):

    largest = my_list[0]
    smallest = my_list[0]
    total = 0

    for nm in num:

        if nm > largest:
            largest = nm

        if nm < smallest:
            smallest = nm

        total = total + nm

    print("Largest number will be: ", largest)
    print("Smallest number will be: ", smallest)
    print("Sum of all numbers will be: ", total)

operations(num)
```

58) Write a Python program to combine values in python list of dictionaries.

300}, o {'item': 'item1', 'amount': 750}]

• Counter ({'item1': 1150, 'item2': 300})

```python
data = [
{'item': 'item1', 'amount': 400},
{'item': 'item2', 'amount': 300},
{'item': 'item1', 'amount': 750}
]

result = {}

for d in data:
    item = d['item']
    amount = d['amount']
    
    if item in result:
        result[item] += amount
    else:
        result[item] = amount

print(result)
```

61) Write a Python function to calculate the factorial of a number (a nonnegative integer)

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

63) Write a Python function to check whether a number is perfect or not.

```python
def is_perfect(num):
    sum_div = 0
    for i in range(1, num):
        if num % i == 0:
            sum_div += i
    return sum_div == num

print(is_perfect(28))
```
