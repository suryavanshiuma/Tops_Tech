Healthcare Industry

Design a Python class ClinicAppointment that manages patient appointments in a clinic. 

The system should have the following features:

➔ Book Appointment:    \
● Prompt for patient name, age, mobile number, and preferred doctor.    \
● Show time slots (10am, 11am, 12pm, 2pm, 3pm).     \
● Check slot availability and confirm booking.

➔ View/Cancel Appointment:     \
● Allow patient to view or cancel their appointment using mobile number.

➔ Doctor Availability:     \
● Maintain a maximum of 3 appointments per time slot per doctor.

➔ Data Persistence:    \
● Store appointments in memory only (no files/dbs required).    

```python
class ClinicAppointment:

    def bkapntmnt(self,pname,page,pcontact):
        self.pname=pname
        self.page=page
        self.pcontact=pcontact

    def prntdata(self):
        print("Patient Name: ",self.pname)
        print("Patient Age: ",self.page)
        print("Contact Details of Patient: ",self.pcontact)

    def slot(self):
        slt_lst=["10am","11am","12pm","2pm","3pm"]
        print("Available SLots: ")
        for i in slt_lst:
            print(i)

        while True:
            selected_slot=input("Please Select Your Preffered Time Slot: ")
            if selected_slot in slt_lst:
                print("Selected Time Slot is ",selected_slot)
                print("Your Appointment has been Successfully Confirmed. Thank You!")
                return selected_slot
            else:
                print("Invalid Slot Selected! Please try again.")


print("Stored Appointments: ")
appointments={}     #dictionary to store data

ca=ClinicAppointment()

nam=input("Enter Name of the Patient: ")
ag=int(input("Enter Age of the Patient: "))

while True:
    contact=input("Enter Mobile Number of Patient: ")

    if contact.isdigit() and len(contact) == 10:
        break
    else:
        print("Invalid Mobile Number! Please enter a 10-digit number")

ca.bkapntmnt(nam,ag,contact)
ca.prntdata()

while True:
    print("Available Doctors:")
    print("1.Dr.Jason")
    print("2.Dr.Maxtern")
    print("3.Dr.Michalle")

    choice= int(input("Enter Your Choice: "))
    print("-"*40)

    if choice==1:
        print("Doctor Selected: Dr.Jason")
        slot_selected = ca.slot()

        appointments[nam] = {
            "Age": ag,
            "Contact": contact,
            "Doctor": "Dr.Jason",
            "Slot": slot_selected
        }
        break
    
    elif choice==2:
        print("Doctor Selected: Dr.Maxtern")
        slot_selected = ca.slot()

        appointments[nam] = {
            "Age": ag,
            "Contact": contact,
            "Doctor": "Dr.Maxtern",
            "Slot": slot_selected
        }
        break
    elif choice==3:
        print("Doctor Selected: Dr.Michalle")
        slot_selected = ca.slot()

        appointments[nam] = {
            "Age": ag,
            "Contact": contact,
            "Doctor": "Dr.Michalle",
            "Slot": slot_selected
        }

        break
    else:
        print("Invalid Choice. Please Try Again!")
        
print(appointments)
```
