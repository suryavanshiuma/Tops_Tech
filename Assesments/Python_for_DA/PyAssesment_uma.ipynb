Healthcare Industry

Design a Python class ClinicAppointment that manages patient appointments in a clinic. 

The system should have the following features:

➔ Book Appointment:    \
● Prompt for patient name, age, mobile number, and preferred doctor.    \
● Show time slots (10am, 11am, 12pm, 2pm, 3pm).     \
● Check slot availability and confirm booking.

➔ View/Cancel Appointment:     \
● Allow patient to view or cancel their appointment using mobile number.

➔ Doctor Availability:     \
● Maintain a maximum of 3 appointments per time slot per doctor.

➔ Data Persistence:    \
● Store appointments in memory only (no files/dbs required).    

```python
class ClinicAppointment:

    def bkapntmnt(self,pname,page,pcontact):
        self.pname=pname
        self.page=page
        self.pcontact=pcontact

    def prntdata(self):
        print("Patient Name: ",self.pname)
        print("Patient Age: ",self.page)
        print("Contact Details of Patient: ",self.pcontact)

    def slot(self):
        slt_lst=["10am","11am","12pm","2pm","3pm"]
        print("Available SLots: ")
        for i in slt_lst:
            print(i)

        while True:
            selected_slot=input("Please Select Your Preffered Time Slot: ")
            if selected_slot in slt_lst:
                print("Selected Time Slot is ",selected_slot)
                print("Your Appointment has been Successfully Confirmed. Thank You!")
                return selected_slot
            else:
                print("Invalid Slot Selected! Please try again.")


print("Stored Appointments: ")
appointments={}     #dictionary to store data

ca=ClinicAppointment()

nam=input("Enter Name of the Patient: ")
ag=int(input("Enter Age of the Patient: "))

while True:
    contact=input("Enter Mobile Number of Patient: ")

    if contact.isdigit() and len(contact) == 10:
        break
    else:
        print("Invalid Mobile Number! Please enter a 10-digit number")

ca.bkapntmnt(nam,ag,contact)
ca.prntdata()

while True:
    print("Available Doctors:")
    print("1.Dr.Jason")
    print("2.Dr.Maxtern")
    print("3.Dr.Michalle")

    choice= int(input("Enter Your Choice: "))
    print("-"*40)

    if choice==1:
        print("Doctor Selected: Dr.Jason")
        slot_selected = ca.slot()

        appointments[nam] = {
            "Age": ag,
            "Contact": contact,
            "Doctor": "Dr.Jason",
            "Slot": slot_selected
        }
        break
    
    elif choice==2:
        print("Doctor Selected: Dr.Maxtern")
        slot_selected = ca.slot()

        appointments[nam] = {
            "Age": ag,
            "Contact": contact,
            "Doctor": "Dr.Maxtern",
            "Slot": slot_selected
        }
        break
    elif choice==3:
        print("Doctor Selected: Dr.Michalle")
        slot_selected = ca.slot()

        appointments[nam] = {
            "Age": ag,
            "Contact": contact,
            "Doctor": "Dr.Michalle",
            "Slot": slot_selected
        }

        break
    else:
        print("Invalid Choice. Please Try Again!")
        
print(appointments)
```

School Management System 

Design a Python class SchoolManagement that helps manage student admissions and 
records.

The system should support:

➔ New Admission:   \
● Collect student name, age, class (1–12), and guardian's mobile number.    \
● Assign a unique student ID automatically.     \
● Validate age: must be between 5 and 18.   \
● Validate mobile number: must be 10 digits.

➔ View Student Details:    \
● Allow lookup using student ID.

➔ Update Student Info:     \
● Update mobile number or class. 

➔ Remove Student Record:   \
● Remove a student using their student ID. 

➔ Exit System

```python
import random

class schoolsystem:

  try:

    def __init__(self):
      self.student_details = {}
      self.menu()

    def menu(self):
        user_input = int(input("""how would you like to proceed?
        1. Enter 1 to New admission
        2. Enter 2 to view student
        3. Enter 3 to update student informatiion
        4. enter 4 to remove student record
        5. Enter 5 to exit
        """))
        if user_input == 1:
          self.new_admission()
        elif user_input == 2:
          self.view_student()
        elif user_input == 3:
          self.update_student()
        elif user_input == 4:
          self.remove_student()
        elif user_input == 5:
            exit()
        else:
            print("Invalid choice. Try again.")
            self.menu()

    def new_admission(self):
          name = input("Enter name: ")
          age = int(input("Enter age: "))
          if age >= 20 or age <= 5:
              print("Invalid age.")
              return
          class_no = int(input("Enter class: 1 to 12 "))
          if class_no >= 12 or class_no <= 1:
              print("Invalid class.")
              return
          mobile = input("Enter mobile number: ")
          if len(mobile) != 10 or not mobile.isdigit():
              print("Invalid mobile number.")
              return
          for student in self.student_details.values():
            if student['mobile'] == mobile:
                print("Student already exists.")
                return

          student_id = random.randint(100000, 999999)
          print(f"your student_id",student_id)

          self.student_details[student_id] = {'name': name, 'age': age, 'mobile': mobile}
          print("Admission successful!")
          print(self.student_details)
          self.menu()

    def view_student(self):
          student_id = int(input("Enter student id: "))
          if student_id in self.student_details:
            print(self.student_details[student_id])
            self.menu()
          else:
            print("Student not found.")
            self.menu()

    def update_student(self):
          student_id = int(input("Enter student id: "))
          if student_id in self.student_details:
            name = input("Enter name: ")
            age = int(input("Enter age: "))
            mobile = input("Enter mobile number: ")
            if len(mobile) != 10 or not mobile.isdigit():
              print("Invalid mobile number.")
              return
            self.student_details[student_id] = {'name': name, 'age': age, 'mobile': mobile}
            print("Student updated successfully!")
            print(self.student_details)
            self.menu()
          else:
            print("Student not found.")
            self.menu()

    def remove_student(self):
          student_id = int(input("Enter student id: "))
          if student_id in self.student_details:
            del self.student_details[student_id]
            print("Student removed successfully!")
            print(self.student_details)
            self.menu()
          else:
            print("Student not found.")
            self.menu()
  except Exception as e:
      print(e)

schoolsystem()
```
